```
df['Individuals Affected'] = pd.to_numeric(df['Individuals Affected'],errors='coerce')
df['Individuals Affected'] = df['Individuals Affected'].replace(np.nan, 0, regex=True)
df['Individuals Affected'] = df['Individuals Affected'].astype(int)
```

```
df.info()
```

```
df.isnull().sum()
```

```
df=df.drop(['Web Description','Name of Covered Entity'], axis = 1)
```

```
df[df.duplicated(keep = False)]
```

```
df=df.drop_duplicates()
df
```

```
plt.figure(figsize=(10,10))
sns.heatmap(df.isnull(), cbar = False, cmap = 'YlGnBu')
```

```
df['State'].value_counts()
```

```
plt.figure(figsize=(15,15))
sns.countplot(x='State',data=df)
```

```
df['Covered Entity Type'].value_counts()
```

```
plt.title("Business Associate Presence")
plt.xlabel("Business Associate Present")
plt.ylabel("Count")
plt.index=['Yes','No']
sns.histplot(df, x="Business Associate Present",shrink=0.4)
```

```
df["Covered Entity Type"].value_counts()
```

```
plt.title("Industries with most breaches")
plt.xlabel("Covered Entity Type")
plt.ylabel("Count")
sns.countplot(x='Covered Entity Type',data=df)
```

```
plt.figure(figsize=(10,10)),
sns.countplot(y='Type of Breach',data=df,order=df['Type of Breach'].value_counts().index)
for index, value in enumerate(df['Type of Breach'].value_counts()):
    plt.text(value, index, str(value))
```

```
plt.figure(figsize=(20,20))
sns.countplot(y= 'Location of Breached Information', data = df, order = df['Location of Breached Information'].value_counts().index)
for index, value in enumerate(df['Location of Breached Information'].value_counts()):
    plt.text(value, index, str(value))
```

```
plt.figure(figsize=(10,10))
gph=sns.catplot(x='Covered Entity Type',hue='Business Associate Present',kind='count',data=df)
for axes in gph.axes.flat:
    _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=45)
```
